# Sniega modelis (Izcelsme - METUL 1993)

METUL ir izveidots no trim konceptuālām daļām, kur pirmā daļā no meteodatiem tiek aprēķinājis sniega segas veidošanos "punktā". Tā kā punktam nav laukuma, tad pieņemts ūdens apjomu, kas ieņemtu trīsdimensionālu telpu, raksturot tikai vienā no trim dimensijām - augstuma dimensijā milimetros $mm$. Līdzīgi tiek mērīti nokrišņi.  Šo fizikālo mēru $mm$ reizinot ar laukumu $mm^2$  var iegūt tilpumu $mm^3$.
Aprēķinos izmantotie meteodati sevī ietver trīs parametrus - nokrišņus $mm$, temperatūru $C^{\circ}$ un mitruma deficītu $g/m^3$
Svarīgi pieminēt, ka temperatūra un mitruma deficīts ir dienas vidējās vērtības, savukārt nokrišņi - dienas summa.

Viena aprēķina ietvaros tiek norādīti sākuma nosacījumi, meteodati un kā rezultāts iegūtas ūdens daudzuma izmaiņas vienas dienas ietvaros. Sākuma nosacījumi norāda, kāda ir bijusi sniega sega "dienas sākumā", meteodati norāda, kādi ir bijuši "dienas apstākļi" un rezultāts raksturo ūdens bilanci "dienas beigās".

## Kas ar ūdeni var notikt sniegā?
* Pirmkārt pie zemām temperatūrām ūdens var uzkrāties sniega formām.
* Otrkārt sniegs atbrīvo ūdeni šķidrā formā sniegam kūstot.
* Treškārt ūdens var iztvaikot pārejot no šķidrā agregātstāvokļa gāzveida agregātstāvoklī.
* Ceturtkārt ūdens var iztvaikot sublimējoties - pārejot no cietā agregātstāvokļa gāzveida agregātstāvoklī.
* Piektārkt ūdens, kas ir izkusis no sniega var atkārtoti sasalt, ja temperatūra pazeminās.

# Kas nosaka secību šiem procesiem modeļa ietvaros?

Pirmkārt to nosaka dienas vidējā temperatūra. 
* Ja dienas vidējā temperatūra ir augstāka par $0 C^{\circ}$, nokrišņu uzkrāšanās sniega segā nenotiks, sniegs kusīs, ūdens iztvaikos, izkusušais ūdens, kas nespēs palikt sniegā un nebūs iztvaikojis, veidos virszemes noteci. 

* Ja temperatūra ir zemāka par $0 C^{\circ}$, nokrišņi uzkrāsies sniega segā, ūdens apjoms tajā pieaugs, sniega segā esošais ūdens šķidrā formā pārvērtīsies sniega/ledus formā. Atmosfērā esošais mitruma deficīts liks daļai no ūdens sniega segā iztvaikot. Arī šeit var veidoties notece, ja viss ūdens šķidrā formā nepārvēršas cietā formā un sniega sega nav spējīga to noturēt sevī.

* Taču ja dienas vidējā temperatūra ir ap $0 C^{\circ}$, tad norisināsies gan nokrišņu uzkrāšanās sniega segā cietā formā, gan arī ūdens kušana. Kā arī iztvaikošana. Tā kā norisinās sniega kušana, var veidoties notece, ja izkūst vairāk ūdens, nekā sniega sega sevī to spēj saturēt.

# Sniega kušana
Aplūkosim katru no galvenajiem procesiem, kas ar ūdeni var notikt, mazliet sīkāk.

Pirmkārt, ja sniega sega ir, tad, atkarībā no temperatūras, dienas laikā no tās radīsies ūdens, kurš iztvaikos vai atkārtoti sasals. Tāpēc kušanu, ja tāda norisināsies, modeļa aprēķinā būs jāveic kā pirmo.

### Mainīgo nosaukumi
Šajā koda daļā ir dažādi mainīgo vērtību nosaukumi.

* ```T```: dienas vidējā temperatūra, $ C^{\circ}$
* ```T2```: temperatūra, virs kuras nokrišņi neuzkrāsies sniega segā cietā formā, $ C^{\circ}$.
* ```SSS```: ūdens cietā formā sniega segā, $mm$.  (Sniega Sega "Solid") 
* ```SSL```: ūdens šķidrā formā sniega segā, $mm$.  (Sniega Sega "Liquid") 
* ```WMELT```: maksimālais ūdens apjoms no sniega, kas var izkust, $mm$.
* ```Cmelt```: koeficients, kurš raksturo sniega kušanas intensitāti dienā uz gŗādu, $mm/C^{\circ}$.

Sniega kušanas daļā loģika balstīta uz to, ka kušana norisināsies, ja ūdens cietā formā vispār būs pieejams kušanas procesam.
Lai aprēķinātu to daļu, kas pārvērtīsies ūdenī, ir nepieciešami sākuma nosacījumi - cik vispār sniega sega cietā formā ```SSS``` ir, cik vispār jau ūdens sniega segā šķidrā formā ```SSL``` ir.

Sākumā Ūdens bilances aprēķinā ir jānoskaidro, cik daudz ūdens no sniega vispār var pārvērsties šķidrā formā ```WMELT```. Tiek pieņemts, ka tā ir lineāra sakarība, kur potenciāli maksimālais sniega daudzums, kurš var pārvērsties šķidrā formā, ir lineāri atkarīgs no dienas vidējās temperatūras un "atskaites temperatūras" starpības.

Ja sniega segā cietā formā ūdens ir vairāk, nekā tas var izkust, ūdens daudzums sniega segā samazināsies par visu ```WMELT``` apjomu, savukārt ūdens daudzums šķidrā formā ```SSL``` palielināsies par visu ```WMELT``` apjomu.

Savukārt, ja potenciālais ūdens daudzums, kas varētu izkust ir lielāks nekā ūdens daudzums sniegā, viss esošais sniegs izkusīs (arī tad, ja tā apjoms sniegā ir mazāks, nekā potenciālais ūdens daudzums ```WMAX```) un pārvērtīsies šķidrā formā ```SSL```

### Izmaiņas, kas ieviestas METUL un METQ versijās, kas veidotas pēc 1993. gada

Pētot A.Zīverta mantojumu, esmu nonācis pie secinājuma, ka citās versijās šajā bloka daļā ir uzlabojumi. Sniega kušanas process ir arī atkarīgs no Saules insolācijas vērtības un tai piesaistītā parametra ```AMELT```. Ja Decembrī un Aprīlī sakrīt modelī izmantotā temperatūra, ir jāņem vērā, ka Aprīlī kušana būs intensīvāka, jo Saules ietekme uz sniega kušanu Aprīlī ir ilgāka - Aprīlī diena ir garāka.




````python
def snowMelt(T,SSS,SSL,T2,Cmelt):
    WMELT=Cmelt*(T-T2)
    
    if(SSS>WMELT):
        SSS=SSS-WMELT
        SSL=SSL+WMELT
    else:
        WMELT=SSS # max reālais sniegs, kas var izkust ir mazāks par potenciālo, kas var izkust
        SSS=0
        SSL=SSL+WMELT
        
        
    return [SSS,SSL,WMELT]
````

# Iztvaikošana

Ja atmosfērā pastāv mitruma deficīts, norisināsies iztvaikošana un sublimācija. Tie ir termodinamiski sarežģīti procesi, kur maksimālo iespējamo ūdens apjomu, kas nokļūs gaisā, galvenokārt var vispārināti noteikt zinot gaisa relatīvo mitrumu un temperatūru. Precīzāka apjoma noteikšanai būtu jāzina arī atmosfēras spiediens. 

### Mainīgo nosaukumi
Šajā koda daļā ir dažādi mainīgo vērtību nosaukumi.
* ```SSS```: ūdens cietā formā sniega segā, $mm$.  (Sniega Sega "Solid") 
* ```SSL```: ūdens šķidrā formā sniega segā, $mm$.  (Sniega Sega "Liquid") 
* ```Def```: absolūtais mitruma deficīts
* ```KS```: koeficients, kurš raksturo iztvaikošanu no sniega. Vajadzētu būt vērtībās starp 0 (neiztvaiko) un 1 (iztvaiko viss)
* ```maxEv```: aprēķinātais maksimālais iespējamais ūdens daudzums no sniega kā absolūtā mitruma deficīta un KS reizinājums, $mm$

Iztvaikošanas daļas loģika ir balstīta uz pieņēmumu, ka ūdens šķidrā formā iztvaiko pirmais, bet tikai tad, kad ūdens šķidrā formā vairāk nav, norisināsies sublimācijas process.

Ja ūdens šķidrā formā ir pieejams vairāk, nekā atmosfēra var paņemt, tad no šķidrās formas tiks atņemts iztvaikojošais ūdens daudzums, pārējais ūdens paliek sniega segā.

Savukārt, ja atmosfēra var paņemt vairāk, nekā tas ir pieejams gan cietā, gan šķidrā formā, tad pieņems, ka iztvaikos viss pieejamais ūdens gan cietā, gan šķidrā formā. Pēc iztvaikošanas ūdens daudzums cietā un šķidrā formā būs 0.

Bet, ja atmosfēra var paņemt vairāk ūdeni, nekā tas pieejams šķidrā formā, bet ūdens cietā formā ir pārāk daudz, tad modelī tiek pieņemts, ka noteikti iztvaikos ūdens šķidrā formā un tikai to daļu, kas iztvaikošanā pietrūkst, paņems no sniega segas cietās formas.

Svarīgi šajā brīdī saprast, ka šī ir tikai un vienīgi modeļa daļa, kas raksturo iztvaikošanu. To nekorekti interpretējot, atraujot no kopējā modeļa loģikas (aprakstīta vēlāk), var nokļūt pie nekorektiem secinājumiem.



````python
def evaporation(Def,SSS,SSL,KS):
    #print(Def,SSS,SSL,KS)
    
        
    maxEv=KS*Def   #maksimālais iespējamais ūdens daudzums iztvaikojot mm aprēķināms ar koeficienta un mitruma deficīta reizinājumu. 
    #ja sniegā ir ūdens, tas iztvaikos vispirms.
    #pēc tam iztvaikos sniegs.

    if (SSL>=maxEv): #ja sniegā vairāk ūdens šķidrā veidā, nekā var max iztvaikot, tad
        SSL=SSL-maxEv # tad ūdens sniegā samazinās par max evaporācijas daļu. vairāk šajā ciklā šajā nosacījumā iztvaikošanas nebūs. Bet
        ev=maxEv
    elif (SSS+SSL-maxEv<0):
        ev=(SSS+SSL)
        SSS=0
        SSL=0       
    else:
        SSS=SSS+SSL-maxEv  #bet ja var iztvaikot vairāk, nekā ir ūdens, tad ūdens iztvaikos viss un pārējais iztvaikos no sniega.
        SSL=0     
        ev=maxEv
    return [SSS,SSL,ev]
````

# Atkārtota ūdens sasalšana sniega segā

Ja dienas vidējā temperatūra būs zem $ C^{\circ}$, šķidrā formā esošais ūdens ```SSL``` pārvērtīsies cietā formā, uzkrāsies sniega segā ```SSS```. Taču tas, cik daudz no tā pārvērtīsies, ir atkarīgs no tā, cik zema būs temperatūra. Jo zemāka temperatūra, jo straujāk, izteiktāk norisināsies ūdens sasalšana.

### Mainīgo nosaukumi
Šajā koda daļā ir dažādi mainīgo vērtību nosaukumi.

* ```T```: dienas vidējā temperatūra, $ C^{\circ}$
* ```T2```: temperatūra, virs kuras nokrišņi neuzkrāsies sniega segā cietā formā, $ C^{\circ}$.
* ```SSS```: ūdens cietā formā sniega segā, $mm$.  (Sniega Sega "Solid") 
* ```SSL```: ūdens šķidrā formā sniega segā, $mm$.  (Sniega Sega "Liquid") 
* ```WMELT```: maksimālais ūdens apjoms no sniega, kas var izkust, $mm$.
* ```Cmelt```: koeficients, kurš raksturo sniega kušanas intensitāti dienā uz gŗādu, $mm/C^{\circ}$.
* ```CFR```: koeficients, kurš raksturo to, kā sniega segā sasalst ūdens.
* ```WRFR```: potenciālais ūdens daudzums, kas varētu sasalt, $mm$.

Līdzīgi kā iztvaikošanas daļā, arī šeit vispirms tiek aprēķināts potenciālais ūdens daudzums, kas varētu sasalt. Ja pieejamais ūdens daudzums ```SSL``` ir mazāk nekā potenciālais ūdens daudzums ```WRFR```, viss pieejamais ūdens daudzums sasals un papildinās ūdens daudzumu cietā formā ```SSS```.

Bet, ja gadījumā ūdens daudzums sniega segā šķidrā formā ir vairāk, nekā var sasalt, tad sasals tikai daļa no ūdens šķidrā formā, savukārt nesasalusī daļa paliks kā ūdens šķidrā formā, mazāk, nekā dienas sākumā.



````python
def refreeze(T,SSS,SSL,T2,CMelt,CFR):
    WRFR = CMelt *  ( T2 - T )  * CFR # maksimālā ūdens daļa, kura sasals.
            
    #ja ūdens šķidrā formā ir mazāk, nekā tas, kas var sasalt, tad viss ūdens sasalst un tas jāpieskaita pie SSS
    if (SSL<WRFR):
        SSS=SSS+SSL  #šķidrums sasalis, jāpieskaita pie cietās fāzes
        SSL=0
    else:
        SSL=SSL-WRFR #aprēķinam, cik paliek pāri pēc sasalšanas
        SSS=SSS+WRFR
        
    return [SSS,SSL]
````

# Sniega segas veidotā notece
Noteci šajā modelī veido tā ūdens šķidrā formas daļa, kuru sniegs nav spējīgs sevī noturēt. Sniegam kūstot daļa no ūdens paliek pašā sniegā un daļa no tā veido noteci.

### Mainīgo nosaukumi
Šajā koda daļā ir dažādi mainīgo vērtību nosaukumi.

* ```SSS```: ūdens cietā formā sniega segā, $mm$.  (Sniega Sega "Solid") 
* ```SSL```: ūdens šķidrā formā sniega segā, $mm$.  (Sniega Sega "Liquid") 
* ```WHT```: potenciālais ūdens daudzums, kas var palikt sniegā, $mm$,
* ```WHC```: koeficients, kurš raksturo cik daudz ūdens, $mm$, var palikt sniega segā, $mm$. 
* ```RS```: ūdens daudzums, kas veido noteci, $mm$.

Lai aprēķinātu, cik daudz ūdens šķidrā formā var palikt sniegā, ir nepieciešams zināt, cik daudz ir pats sniegs. Tā kā sniegs arī ir ūdens, bet cietā formā un tā blīvums ir mazāks par ūdens blīvumu šķidrā formā, tad var teikt, ka ```WHT``` ir ūdens šķidrā formā iekš tā cietā formas. Sniegs sevī notur ūdeni. Viens milimetrs sniegs sevī var turēt kādu mazu daļu no viena milimetra ūdens. Koeficients ```WHC``` nevar būt negatīvs, jo tad tas nozīmētu, ka sniegs no sevis ūdeni izspiež. Ja koeficients ir vienāds ar ```0```, tad tāds koeficients raksturo sausu sniegu. Savukārt ļoti slapju sniegu raksturo koeficients ```0.08 - 0.15```. Atsauce šeit - https://www.hec.usace.army.mil/confluence/hmsdocs/hmstrm/snow-accumulation-and-melt/snow-properties

Šeit ir saskatāms modeļa uzlabojums, ka koeficients ```WHC``` sniega segai veidojoties ir mainīgs un atkarīgs no izkusušā un atkārtoti sasalstošā sniega daudzuma. Sākuma nosacījums ```WHC``` vērtībai ir pētāms un atkarīgs no temperatūras. Izteikti zemā temperatūrā tā vērtībai būtu jābūt vienādai ar 0, taču temperatūrai pieaugot un sniegam kūstot, tā vērtībai jāmainās līdz ar ūdens šķidrās formas izmaiņām.


````python
def snowRunoff(SSS,SSL,WHC):
    WHT=SSS*WHC #maksimālā iespējamā ūdens daļa, kas var palikt sniegā.
            
    if (SSL>WHT): #ja šķidrā formā ūdens ir vairāk, nekā tā daļa, kas var palikt sniegā, tad notece veidosies no starpības.
        RS=SSL-WHT
        SSL=SSL-RS
    else:
        RS=0     # ja šķidrā formā ūdens ir mazāk, nekā sniegs spēj noturēt, tad notece neveidosies un ūdens paliks sniegā.
        
    return [SSS, SSL, RS]
````

# Sniega segas modeļa loģika

Tā kā to, kas notiks ar ūdeni sniega segā nosaka meteoroloģiskie dati, tad loģika jāveido balstoties uz tiem. Pirmkārt kā zarojuma nosacījums ir izvēlēta temperatūra, jo no tās atkarīgi tālākie procesi. Otrkārt jānoskaidro, vai eksistē nokrišņi, kas ūdens saturu sniega segā papildina un treškārt - cik un kā ūdens no sniega segas iztvaikos.

* Ja temperatūra ir lielāka par kādu noteiktu kritisko temperatūras vērtību ```T1```, sniegs kusīs; bet pirmkārt, ūdens saturs sniegā ```SSL``` tiks papildināts ar nokrišņiem ```P``` šķidrā formā. Šķidro formu papildinās arī tā daļa, kas paralēli no sniega kusīs ```WMELT```. Ūdens saturs no sniega segas arī iztvaikos, balstoties uz iepriekšaprakstīto pieņēmumu - pimkārt iztvaikos ūdens šķidrā formā. Pāri palikušais ūdens, ja tāds būs, veidos noteci ```RS```.

* Savukārt, ja dienas vidējā temperatūra ir ap nulli grādiem, jeb matemātiski starp divām robežvērtībām - augstāko temperatūras robežvērtību ```T1``` un zemāko temperatūras robežvērtību ```T2```, tad nokrišņi uzkrāsies sniega segā cietā formā ```SSS```, bet norisināsies arī sniega kušanas process, bet ne sasalšanas process. Ūdens sniega segas cietā formā  ```SSS``` papildinās par klāt pienākušo nokrišņu apjomu, bet ūdens sniega segā šķidrā formā ```SSL``` papildinās par ```WMELT``` apjomu. Norisinās arī iztvaikošana un ```SSL``` samazinās. Pāri palikušais ūdens šķidrā formā ```SSL``` var samazināties un veidot noteci ```RS```.

* Bet ja dienas vidējā temperatūra ir zemāka par zemāko temperatūras robežvērtību ```T2``` (ir sals), nokrišņi ```P```, ja tādi ir, papildina ūdeni cietā formā sniega segā ```SSS```, nekas nekūst, bet joprojām norisinās iztvaikošana. Ja sniega segā ir ūdens šķidrā formā ```SSL```, tā apjoms samazināsies pilnīgi vai daļēji sasalstot un palielinot ```SSS``` apjomu. Savukārt tā daļa, kas nesasalst, potenciāli var veidot noteci ```RS```.

````python
def snowmodel(T,P,Def,SSS,SSL,KS,T1,T2,CMelt,WHC,CFR):
    if T>=T1:
        #kūst
        #print("temp ir lielāka par ", T1)
        SSL=SSL+P
        #print ("nokrišņi ", P, " mm ir šķidrā formā un papildina ūdens daudzumu sniegā ", SSL, "mm")
        SSS, SSL, WMELT = snowMelt(T,SSS,SSL,T2,CMelt)
        #print ("sniegs kūst, tāpēc sniegs cietā formā samazinās par ", WMELT, "mm")
        #print ("pēc kušanas sniega segā ir {0:3.2f} mm".format(SSS))
        #print ("pēc kušanas ūdens šķidrā formā sniega segā ir ",SSL, " mm")
        
        #iztvaiko
        SSS, SSL, ev = evaporation(Def,SSS,SSL,KS)
        #print ("ir iztvaikojis ", ev , " mm")
        #print ("pēc iztvaikošanas sniega segā sniegs ir ", SSS, " mm")
        #print ("pēc iztvaikošanas sniega segā ūdens ir ", SSL, " mm")
        
        #notek
        
        if (SSL>0):
            SSS, SSL, RS = snowRunoff(SSS,SSL,WHC) #daļa no ūdens veidos noteci. SSL samazināsies par RS daļu.
        else:
            RS=0
        
    elif T2<T<T1:
        #daļa sasalst, daļa iztvaiko
        #print("temp ir starp ",T1, " un ", T2)
        

        # RS=WMelt
        # SSS = SSS+P-RS-ES
        
        
        #print ("uzkrājas nokrišņi P: ",P, "mm, cietā formā un nāk klāt sniega segā esošajiem: ", SSS, "mm" )
        #print ("Sniegs cietā formā sniega segā ir ", P+SSS, " mm")
        #print ("Sniegs šķidrā formā sniega segā ir ", SSL, " mm" )
        SSS=SSS+P
        
        SSS, SSL, WMELT = snowMelt(T,SSS,SSL,T2,CMelt)
        #print ("sniegs kūst, tāpēc sniega sega samazinās par ", WMELT, " mm")
        #print ("pēc kušanas sniega segā ir {0:3.2f} mm".format(SSS))
        #print ("pēc kušanas ūdens šķidrā formā sniega segā ir ",SSL, " mm")
        
        SSS, SSL, ev = evaporation(Def,SSS,SSL,KS)
        #print ("ir iztvaikojis ", ev , " mm")
        #print ("pēc iztvaikošanas sniega segā sniegs ir ", SSS, " mm")
        #print ("pēc iztvaikošanas sniega segā ūdens ir ", SSL, " mm")
        
       
        
        if (SSL>0):
            SSS, SSL, RS = snowRunoff(SSS,SSL,WHC) #daļa no ūdens veidos noteci.
        else:
            RS=0
        
    elif T<=T2:
        #print("temp ir zem ", T2)
        # nokrišņi veido sniega segu. ja ir mitruma deficīts, ūdens no sniega iztvaiko.
        # nokrišņi uzkrājas cietā formā sniega segā "SSS" Sniega Sega Solid
        SSS=SSS+P
        SSS, SSL, ev = evaporation(Def,SSS,SSL,KS)
        # ja sniegā ir ūdens, pēc iztvaikošanas jāaprēķina tā daļa, kura sasals un kura pārtecēs    
        if (SSL>0):   # ja sniega segā ir ūdens
            SSS, SSL = refreeze(T,SSS,SSL,T2,CMelt,CFR) #daļa no tā sasals un veidos mazāk šķidra ūdeni sniegā un vairāk sniegu cietā formā
            SSS, SSL, RS = snowRunoff(SSS,SSL,WHC) #daļa no ūdens veidos noteci.
        else:
            RS=0

    #print("SSS={0:3.2f}, SSL={1:3.2f}, ev={2:3.2f}, RS={3:3.2f}, summa={4:3.2f}".format(SSS,SSL,ev,RS,sum([SSL,SSS,ev,RS])))
    return [SSS,SSL,ev,RS,sum([SSL,SSS,ev,RS])]
        
````
